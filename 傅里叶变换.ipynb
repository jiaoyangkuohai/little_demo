```python
import numpy as np
import cmath
%matplotlib inline
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
```

```python
np.random.seed(0)
img = np.random.randint(0,10, (3,3))
img
```

```python
fft_img = np.fft.fft2(img)
fft_img
```
