```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
%config InlineBackend.figure_format = 'retina' # 视网膜技术
```

```python
def set_seed(random_seed=2):
    random.seed(random_seed)
    np.random.seed(random_seed)
```

```python
img = np.asarray(range(16), dtype=np.float32).reshape(4,4)
img
```

```python
list_img = []
set_seed(1)
for i in range(16):
    list_img.append(random.randint(0,16))
img2 = np.asarray(list_img, dtype=np.float32).reshape(4,4)
img2
```

```python
plt.imshow(img2, 'gray');
```
